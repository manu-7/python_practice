### palindrome


```python
n = 11
is_prime = True

if n<2:
    is_prime = False
else:
    for i in range(2,n):
        if i*i>n:
            break
        if n%1 == 0:
            is_prime = False
            break
print("Prime" if is_prime else "Not Prime")
```

```python
num = int(input("enter a number:"))

if num < 0:
    print("Invalid number")
    
else:
    rev = int(str(num)[::-1])
    
    if num == rev:
        print("Palindrome")
    else:
        print("Not Palindrome")
```

### Prime or not


```python
num = int(input("Enter a number:"))

if num <= 1:
    print("Not PRime")

for i in range(2,int(num**0.5)+1):
    if num % i == 0:
        print("Not Prime")
        break
    else:
        print("Prime")
```
