#### Armstrong number

```python
num = int(input("Enter the number:"))

n = len(str(num))

temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** n
    temp //= 10
    
if num == total:
    print(num," is an armstrong number")
else:
    print(num,"is not an armstrong number")
    
```

#### HCF/GCD

```python
def gcd(a,b):
    while b!=0:
        a, b = b, a%b
    return a

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

result = gcd(x,y)
print("GCD of the given numbers is: ",result)
```

### 2nd largest Element


```python
n = int(input("Enter how many numbers: "))

arr = []
for i in range(n):
    arr.append(int(input()))   
    
    
largest = float('-inf')
second_largest = float('-inf')

for num in arr:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

if second_largest == float('-inf'):
    print("No second largest number")
else:
    print("Second largest:", second_largest)
```

#### Reverse array using recursion

```python
def reverse_array(arr, left, right):
    # Base condition
    if left >= right:
        return
    
    # Swap elements
    arr[left], arr[right] = arr[right], arr[left]

    reverse_array(arr, left + 1, right - 1)

arr = list(map(int, input("Enter array elements: ").split(',')))

reverse_array(arr, 0, len(arr) - 1)

print("Reversed array:", *arr)

    
    
    
```

```python
def palindrome(s, left, right):

    if left >= right:
        return True

    if s[left] != s[right]:
        return False
    
    return palindrome(s, left + 1, right - 1)


s = input("Enter the string: ")

if palindrome(s, 0, len(s) - 1):
    print("palindrome")
else:
    print("not palindrome")
```

```python
S = "13542"

changed = True

while changed:
    changed = False
    i = 0
    new_s = ""

    while i < len(S):
        # Check adjacent pair
        if i + 1 < len(S) and (int(S[i]) + int(S[i + 1])) % 2 == 0:
            changed = True
            i += 2   # skip both digits (removed)
        else:
            new_s += S[i]
            i += 1

    S = new_s

# Final output
print(S if S else -1)
```

### sort array according to frequency

```python
arr = [10,10, 20, 30, 40,40]

freq = {}

for x in arr:
    freq[x] = freq.get(x,0)+1

result = sorted(arr,key=lambda x:freq[x],reverse=True)
print(result)
```
