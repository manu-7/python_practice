### Inverted triangle:

```
* * * *
* * * 
* * 
* 
```

````python
n = 4
for i in range(n):
    for j in range(n-i):
        print("*",end=" ")
    print()
````

### Pyramid:

```

 ***
*****
```

````python
n = 4
for i in range(n):
    print(" "*(n-i-1) + "*" *(2*i+1))
    
````

### Inverted Pyramid:

```
*****
 ***
  *
```

````python
n = 4
for i in range(n,0,-1):
    for j in range(0,n-i):
        print(end=" ")
    for k in range(0,i):
        print("*",end=" ")
    print()
    
````
