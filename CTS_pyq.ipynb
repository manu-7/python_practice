1.Make a list of 26 zeros for each letter.

2.Convert each character to its position using ASCII.

3.Count how many times each letter appears.

4.Loop from 'a' to 'z'.

5.If the letter count is > 0, add it to the result.

6.Print the final compressed string.

```python
s = input().strip()

freq = [0] * 26   # frequency array for a-z

# count frequency
for ch in s:
    freq[ord(ch) - ord('a')] += 1

result = ""

# build output in alphabetical order
for i in range(26):
    if freq[i] > 0:
        result += chr(ord('a') + i)
        result += str(freq[i])

print(result)
```

### Kth smallest in sorted matrix

```python
import heapq

def kthSmallest(matrix,k):
    m = len(matrix)
    n = len(matrix[0])
    
    maxHeap = []
    
    for r in range(m):
        for c in range(n):
            heapq.heappush(maxHeap,-matrix[r][c])
            
            if len(maxHeap) > k:
                heapq.heappop(maxHeap)
                
    return -maxHeap[0]

matrix = [
    [1, 5, 9],
    [10, 11, 13],
    [12, 13, 15]
]
k = 8

print(kthSmallest(matrix, k))
```

⭐35. PROBLEM RECONSTRUCTION

You are given an array A of length N.

Noah must perform:

1. Select all integer pairs (x, y) from the array whose sum = 18

(i.e., x + y = 18)

2. From these pairs, select the pair with:

maximum product (x * y)

and x > y

3. Return that pair as an array [x, y]

If no such pair exists → return empty array.

```python
def findPair(arr):
    s = set(arr)
    maxProduct = -1
    best = []

    for x in arr:
        y = 18 - x
        if y in s and x > y:     # condition: sum = 18 and x > y
            if x * y > maxProduct:
                maxProduct = x * y
                best = [x, y]

    return best


# --- REQUIRED INPUT FORMAT ---
N = int(input())
A = list(map(int, input().split()))

result = findPair(A)
print(result)
```

⭐ Problem Statement

You are given an integer N.

Find the smallest number X such that:

N % X = N


(This happens when X > N)

Compute the sum of digits of X.

Return this sum as the secret code for N.

```python
n = int(input("Enter the number:"))
x = n+1

sum = 0
while x>0:
    digit = x % 10
    sum += digit
    x //= 10
print(sum)

```

A prefix of M is defined as a number formed by taking the first k digits of M, for all possible non-empty values of k.

Example:
If M = 5043, its prefixes are:

5

50

504

5043

The Identifier(M) is defined as the sum of all non-empty prefixes of M.

Your task:

Return Identifier(M).

```python
def identifier(M):
    n = int(input("enter the number"))
    S = str(n)
    total = 0

    for i in range(1, len(S) + 1):
        total += int(S[:i])

    return total


print(identifier(M))   
```
