### Kth smallest in sorted matrix

```python
import heapq

def kthSmallest(matrix,k):
    m = len(matrix)
    n = len(matrix[0])
    
    maxHeap = []
    
    for r in range(m):
        for c in range(n):
            heapq.heappush(maxHeap,-matrix[r][c])
            
            if len(maxHeap) > k:
                heapq.heappop(maxHeap)
                
    return -maxHeap[0]

matrix = [
    [1, 5, 9],
    [10, 11, 13],
    [12, 13, 15]
]
k = 8

print(kthSmallest(matrix, k))
```

### Problem Statement: Qualifying Score

There is a competition in a school. To participate in the competition, a student must obtain a qualifying score of 35 or more.

A student studies N subjects. The marks obtained in Semester 1 and Semester 2 for these subjects are provided in two integer arrays S1 and S2, respectively.

The qualifying score is calculated using the following steps:

For each subject i (0 ‚â§ i < N), calculate the difference between the marks of Semester 2 and Semester 1 as:
S2[i] ‚àí S1[i].

From the calculated differences, select the top P maximum values and compute their sum.
This sum represents the qualifying score.

If the qualifying score is greater than or equal to 35, print

Qualified <score>


Otherwise, print

Disqualified <score>

Input Format

An integer N, representing the number of subjects

An array S1 of N integers, representing Semester 1 marks

An array S2 of N integers, representing Semester 2 marks

An integer P, representing the number of subjects to be considered

```python
n = int(input())

s1 = list(map(int, input().split(',')))
s2 = list(map(int, input().split(',')))

p = int(input())

diff = []

# Calculate differences
for i in range(n):
    diff.append(s2[i] - s1[i])

# Sort in descending order
diff.sort(reverse=True)

# Handle case where p > n
p = min(p, n)

score = 0
for i in range(p):
    score += diff[i]

# Output result
if score >= 35:
    print(f"Qualified {score}")
else:
    print(f"Disqualified {score}")
```

```python
def qualifyingScore(S1, S2, P):
    n = len(S1)

    # Step 1: Calculate differences
    diff = []
    for i in range(n):
        diff.append(S2[i] - S1[i])

    # Step 2: Sort in descending order
    diff.sort(reverse=True)

    # Step 3: Sum top P values
    P = min(P, n)
    score = 0
    for i in range(P):
        score += diff[i]

    # Step 4: Return result
    if score >= 35:
        return "Qualified " + str(score)
    else:
        return "Disqualified " + str(score)
```

```python
def string_swap(s, operations):
    s = list(s)   # convert string to list for easy swapping
    n = len(s)

    for op in operations:
        if op == 1:
            # Swap first and last character
            s[0], s[-1] = s[-1], s[0]

        elif op == 2:
            # Swap first N/2 and last N/2 characters
            half = n // 2
            s = s[half:] + s[:half]

    return "".join(s)

print(s)
```

```python
import math

def maxPermutationCount(input1, input2):
    vowels = set('aeiouAEIOU')
    max_perm = 0

    for word in input1:
       
        count = 0
        for ch in word:
            if ch not in vowels:
                count += 1

        # calculate factorial if count > 0
        if count > 0:
            perm = math.factorial(count)
            max_perm = max(max_perm, perm)

    return max_perm

input1 = ["hello", "ccbc", "aaeiou"]
input2 = 3

print(maxPermutationCount(input1, input2))

```

### Problem Statement

You are given an integer array A of length N.
You need to split the array at some index i (where 0 ‚â§ i < N‚àí1) into two non-empty parts:

Left subarray: elements from index 0 to i

Right subarray: elements from index i+1 to N‚àí1

A split is called fair if the absolute difference between the sum of elements of the left subarray and the sum of elements of the right subarray is an even number.

Your task is to count and return the total number of fair splits possible for the given array.

Input Specification

input1: An integer N, the size of the array

input2: An integer array A of size N

Output Specification

Return an integer representing the number of fair splits.

input:

N = 5

A = [10, 10, 3, 7, 6]

output: 4

```python
def count_fair_splits(N, A):
    total_sum = 0
    for num in A:
        total_sum += num

    count = 0
    prefix_sum = 0

    for i in range(N - 1):
        prefix_sum += A[i]
        suffix_sum = total_sum - prefix_sum

        if abs(prefix_sum - suffix_sum) % 2 == 0:
            count += 1

    return count

N = 5
A = [10, 10, 3, 7, 6]
print(count_fair_splits(N, A))
```

### Magical Library

Problem: In a magical library, each bookshelf is represented by a two-dimensional array A, where each row of the 2D array A[i] represents the series value of a book.

A row is considered magical if the sum of the odd values of the series of a book is even. Your task is to find and return an integer value representing the number of magical rows.

Input Specification:

input1: An integer value representing the number of rows in the 2D array.
input2: An integer value representing the number of columns in the 2D array.
input3: A 2D integer array where each row represents a series of books.
Output Specification:
Return an integer value representing the number of magical rows.

Example 1:

input1: 3

input2: 3

input3: ((1, 2, 3), (4, 5, 6), (7, 8, 9))

Output: 2

```python
def count_magical_rows(A):
    count = 0

    for row in A:
        odd_count = 0
        for val in row:          
            if val % 2 != 0:
                odd_count += 1

        if odd_count % 2 == 0:
            count += 1

    return count


def main():
    A = [
        [1, 2, 3],
        [4, 6, 8],
        [5, 7, 9],
        [2, 4, 6]
    ]

    result = count_magical_rows(A)
    print(result)


if __name__ == "__main__":
    main()

    
```

### Problem statement:
