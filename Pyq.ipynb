```python
Litres = float(input("Enter the quantity of Diesel: "))
Distance = float(input("Enter the distance covered: "))

if Litres <= 0 or Distance <= 0:
    print("is an Invalid Input")
else:
    liters_per_100km = (Litres / Distance) * 100
    
# in gallon 

miles = 0.6214 * Distance
gallon = 0.2642 * Litres

mpg = miles/gallon

print("Liters/100KM")
print(f"{liters_per_100km:.2f}")

print("Miles/gallons")
print(f"{mpg:.2f}")
```

You want to buy a particular stock at its lowest price and sell it later at its highest price. Since the stock market is unpredictable, you steal the price plans of a company for this stock for the next N days.
Find the best price you can get to buy this stock to achieve maximum profit.

Note: The initial price of the stock is 0.

```python
n = int(input("Enter the number of days: "))

arr = list(map(int, input().split()))   # takes: 3 -2 5 -1 4

price = 0
min_price = 0

for change in arr:
    price += change
    min_price = min(min_price, price)

print(min_price)
```

Problem Statement â€“ Chaman planned to choose a four digit lucky number for his car. His lucky numbers are 3,5 and 7. Help him find the number, whose sum is divisible by  3 or 5 or 7. Provide a valid car number, Fails to provide a valid input then display that number is not a valid car number. 

Note : The input other than 4 digit positive number[includes negative and 0] is considered as invalid.

Refer the samples, to read and display the data.

Sample Input 1:

Enter the car no:1234
Sample Output 1:

Lucky Number

```python
num = int(input("Enter the number"))

if num < 1000 or num > 9999:
    print("Invalid number:")

sum = 0
temp = num

while temp > 0:
    digit = temp %10
    sum += digit
    temp //= 10
    
if sum % 5 == 0 or sum % 3 == 0 or sum % 7 == 0:
    print("Lucky Number")
else:
    print("Not lucky number")
```

```python
a = int(input("Enter the first number"))
b = int(input("Enter the second number:"))


if a <= 0 or b <= 0 or a >= b or (b - a) <= 0:
    print("Provide valid input")
    
else:
    for num in range(a,b+1):
        if num <= 1:
            continue
        for i in range(2,int(num ** 0.5)+1):
            if num % i == 0:
                break
        else:
            print(num,end=" ")
```

Problem Summary (Simple Explanation)
We have:

n products

Each product has a weight

A sensor calculates the impact sum of each product

ðŸ‘‰ What is impact sum?
For product at index i

impact[i] = sum of all weights except weight[i]

If array = [1, 2, 3, 4]

For index 0 â†’ impact = 2 + 3 + 4 = 9
For index 1 â†’ impact = 1 + 3 + 4 = 8
For index 2 â†’ impact = 1 + 2 + 4 = 7
For index 3 â†’ impact = 1 + 2 + 3 = 6


```python
n = int(input())
arr = list(map(int, input().split(',')))
threshold = int(input())

total_sum = sum(arr)
arr2 = []

for i in range(n):
    impact_sum = total_sum - arr[i]  

    if impact_sum > threshold:
        arr2.append(threshold)
    else:
        arr2.append(impact_sum)

print(arr2)

    
```
